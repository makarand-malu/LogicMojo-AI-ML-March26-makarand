Q1. Division with Exception Handling

Write a program that takes two numbers and safely performs division.
Handle division by zero using try-except.

````python
def input_two_numbers(a, b):
    try:
        return float(a) / float(b)
    except ZeroDivisionError:
        return "Error: You cannot divide by zero."
    except ValueError:
        return "Error: Please provide valid numbers."
    except Exception as e:
        return f"An unexpected error occurred: {e}"
a= 100
b = 'z'

input_two_numbers(a, b)
````

Q2. String to Integer Conversion

Given a list:

data = ["10", "20", "abc", "30"]

Convert all values to integers.
Skip invalid values using exception handling.

````python
def input_list(X):
    data = []
    for i, item in enumerate(X):
        try:
            # Try to convert and add to our list
            data.append(int(item))
        except ValueError:
            # If it's "abc", we just skip it and move to the next item
            print(f"Skipping invalid value: {item}")
            continue 
    return data
    
data = ["10", "20", "abc", "30", "cda"]
result = input_list(data)
print(f"Final list: {result}")
````

Q. 19 Extract Valid Transaction Records Using a Single Regex Pattern

You are given a messy log string:

log = """
ID:TXN12345 | User: Mahipal_01 | Email: mahi123@gmail.com | Phone: +91-9876543210 | Amount: Rs.5000
ID:TXN12 | User: @InvalidUser | Email: wronggmail.com | Phone: 98765 | Amount: Rs.-200
ID:TXN99999 | User: Ravi_kumar | Email: ravi.kumar@yahoo.com | Phone: +91-9123456789 | Amount: USD300
ID:TXN54321 | User: Sita99 | Email: sita99@gmail | Phone: +91-9999999999 | Amount: Rs.7000


Write a single REGEX pattern that extracts ONLY VALID records where:
CONDITIONS (VERY IMPORTANT)
Transaction ID
Format: TXN followed by exactly 5 digits
Example: TXN12345
Use: ^, {}, \d

Username
Only letters, numbers, underscore
Must start with a letter
Example: Mahipal_01
Use: [a-zA-Z], \w, +

Email
Valid format: word@word.domain
Example: test123@gmail.com
Use: \w, ., @, +, \.

Phone Number
Format: +91-XXXXXXXXXX
Exactly 10 digits
Use: \+, \d, {}

Amount
Currency: Rs or USD
Amount must be positive
Example: Rs.5000, USD300
Use: (rs|usd), \d+, ?




EXPECTED OUTPUT

Only valid records should match:

[
"ID:TXN12345 | User: Mahipal_01 | Email: mahi123@gmail.com | Phone: +91-9876543210 | Amount: Rs.5000",

"ID:TXN99999 | User: Ravi_kumar | Email: ravi.kumar@yahoo.com | Phone: +91-9123456789 | Amount: USD300"
]

````python
log = """ ID:TXN12345 | User: Mahipal_01 | Email: mahi123@gmail.com | Phone: +91-9876543210 | Amount: Rs.5000 ID:TXN12 | User: @InvalidUser | Email: wronggmail.com | Phone: 98765 | Amount: Rs.-200 ID:TXN99999 | User: Ravi_kumar | Email: ravi.kumar@yahoo.com | Phone: +91-9123456789 | Amount: USD300 ID:TXN54321 | User: Sita99 | Email: sita99@gmail | Phone: +91-9999999999 | Amount: Rs.7000
"""
````

````python
records = re.split(r'(?=ID:)', log)
records = [r.strip() for r in records if r.strip()]
# print(records)

for r in records:
    # Validate ID format
    match = re.search(r'ID:TXN(\d{5})\b', r)
    
    if not match:
        continue 

    username = re.search(r'[a-zA-Z]\w+', r)
       
    if not username:
        continue  
    
    email = re.search('\w+@\w+[.]\w+', r)
    
    if not email:
        continue 
        
    phone_number = re.search('\d{10}', r)

    if not phone_number:
        continue 

    amount_currency = re.search('(Rs.|USD)\d+', r)
    
    if not amount_currency:
        continue 
            
    print(r)  
````

## 1. Customer Segmentation Engine

You are given a list of customer dictionaries:

```python
customers = [
    {"id": 1, "age": 25, "income": 50000, "spend": 20000},
    {"id": 2, "age": 45, "income": 120000, "spend": 80000},
    {"id": 3, "age": 35, "income": 70000, "spend": 30000}
]
```

Task:

1. Write a function segment_customer(customer):
2. If spend divided by income is greater than 0.6 return "High Value"
3. If spend divided by income is between 0.3 and 0.6 return "Medium Value"
4. Otherwise return "Low Value"
5. Create a final dictionary in the format {customer_id: segment}

````python
def segment_customer(customer):
    customer_id = {}
    for item in customer:
        if item["income"] <=0:
            print(f"For the id {item["id"]} income is zero or negative")
        if float(item["spend"])/float(item["income"])>0.6:
            customer_id[item["id"]] = 'High Value'
        elif float(item["spend"])/float(item["income"])>=0.3:
            customer_id[item["id"]] = 'Medium Value'
        else:
            customer_id[item["id"]] = 'Low Value'
    return customer_id
            
    
````

````python
customers = [
    {"id": 1, "age": 25, "income": 50000, "spend": 20000},
    {"id": 2, "age": 45, "income": 120000, "spend": 80000},
    {"id": 3, "age": 35, "income": 70000, "spend": 30000}
]
customer_id = segment_customer(customers)
print(customer_id)
````

````python
def is_fraud(transaction):
    if transaction["amount"] > 100000 or transaction["location"] == 'Unknown':
        transaction["category"] = "fraud"
        return transaction
    else: 
        transaction["category"] = "safe"

raw_results = map(is_fraud, transactions)

# 2. Filter removes the None entries
fraud_transactions = list(filter(None, raw_results))

# category_list = ["fraud" if x["amount"] > 100000 or x["location"] == "Unknown"   else "safe" for x in transactions]


print(fraud_transactions)
````

## 5. Custom Reduce Function

You are given:

```python
accuracies = [0.8, 0.75, 0.9, 0.85]
```

Task:

1. Use reduce function to calculate average accuracy
2. Do not use sum function
3. Show intermediate logic using a custom function


````python
accuracies = [0.8, 0.75, 0.9, 0.85]

from functools import reduce

def my_function(accumulator, current_item):
    print(f"Step: {accumulator}, sum is at the value {current_item}")
    return accumulator + current_item

result = reduce(my_function, accuracies)

average_accuracy = result / len(accuracies)
print(average_accuracy)
````

````python
accuracies = [0.8, 0.75, 0.9, 0.85]

# Custom function to track intermediate logic
def sum_and_track(accumulator, current_value):
    print(f"Step: {i},  {accumulator} + {current_value} = {round(accumulator + current_value, 2)}")
    return accumulator + current_value

# Step 1: Use reduce to get the total sum
total_sum = reduce(sum_and_track, accuracies)


# Step 2: Divide by total count to get average
average_accuracy = total_sum / len(accuracies)

print(f"\nFinal Average Accuracy: {average_accuracy}")
````

## 6. Sequence Pattern Detection

You are given:

```python
data = [10, 20, 30, 25, 40, 50]
```

Task:

1. Write a function detect_increasing_sequence(data)
2. Find the longest continuous increasing sequence
3. Return that sequence as a list

Example output:

```python
[10, 20, 30]
```


````python
data = [10, 20, 30, 25, 40, 50]
buckets = [[] for _ in data]
j = 0

for i, item in enumerate(data):
    if i == 0:
        buckets[j].append(item)
    else:
        if item >= data[i-1]: # Check if the sequence continues
            buckets[j].append(item)
        else:
            j += 1 # Move to next bucket
            buckets[j].append(item) # Don't forget to add the item!

# Filter out empty buckets before finding max
result = max([b for b in buckets if b], key=len)
print(result) # Output: [10, 20, 30]
````

````python
#AI version

data = [10, 20, 30, 25, 40, 50]
all_sequences = []
current_seq = []

for item in data:
    # If it's the first item OR it continues the sequence
    if not current_seq or item >= current_seq[-1]:
        current_seq.append(item)
    else:
        # Save the finished sequence and start a new one
        all_sequences.append(current_seq)
        current_seq = [item]

# Add the very last sequence
all_sequences.append(current_seq)

# Get the winner
longest = max(all_sequences, key=len)
print(longest) # Output: [10, 20, 30]
````

## 7. Dictionary-Based Recommendation System

You are given:

```python
user_purchases = {
    "user1": ["apple", "banana"],
    "user2": ["banana", "orange"],
    "user3": ["apple", "orange"]
}
```

Task:

1. Write a function recommend(user)
2. Recommend items based on common items between users
3. Do not recommend items already purchased by the user
4. Use sets and dictionary iteration

---

````python
user_purchases = {
    "user1": ["apple", "banana"],
    "user2": ["banana", "orange"],
    "user3": ["apple", "orange"]
}
````

````python
# Instead of set(user_purchases.values())
# all_items = set()
# for items in user_purchases.values():
#     all_items.update(items)

def recommend(user, user_purchases):
    if user not in user_purchases:
        return []
    all_items = set()
    for items in user_purchases.values():
        all_items.update(items)
    return all_items - set(user_purchases[user])

recommend("user1", user_purchases)
````

## 8. Recursive Decision Tree Simulation

Task:

1. Write a recursive function decision_tree(score)
2. If score is greater than 80 return "Approved"
3. If score is between 50 and 80 call the function again with score plus 10
4. If score is less than 50 return "Rejected"

---


````python
def decision_tree(score):
    if score > 80:
        return "Approved"
    if  50 <= score <= 80:
        return decision_tree(score+10)
    if score < 50:
        return "Rejected"


decision_tree(60)
````
